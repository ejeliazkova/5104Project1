```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import statsmodels.api as sm

# Let's try reading directly from the csv provided
df_csv = pd.read_csv('Concrete_Data.csv')

# Extract specified row locations for TRAINING data samples and drop the DV
train = df_csv.drop(index=range(501, 631))
X_train = train.drop(columns=['Concrete compressive strength(MPa, megapascals) '])
y_train = train['Concrete compressive strength(MPa, megapascals) ']

# Extract specified row locations for TESTING data samples and drop the DV which has a trailing space fun
test = df_csv.iloc[501:631]
X_test = test.drop(columns=['Concrete compressive strength(MPa, megapascals) '])
y_test = test['Concrete compressive strength(MPa, megapascals) ']

# Standardize using training set mean
# & standard deviation like in Part B
X_train_mean = X_train.mean(axis=0)
X_train_std = X_train.std(axis=0)
X_train = (X_train - X_train_mean) / X_train_std
X_test = (X_test - X_train_mean) / X_train_std

X_train = sm.add_constant(X_train)
X_test = sm.add_constant(X_test)

# Fit OLS model on TRAINING data
model_train = sm.OLS(y_train, X_train)
result_train = model_train.fit()
print(result_train.summary())
print("\n\nMSE of training data:")
print(result_train.mse_resid)

print("\nP values of training data:")
print(result_train.pvalues)

# Predict on TESTING data using the TRAINING model
y_pred_test = result_train.predict(X_test)
mse_test = np.mean((y_test - y_pred_test) ** 2)
ss_res = np.sum((y_test - y_pred_test) ** 2)
ss_tot = np.sum((y_test - np.mean(y_test)) ** 2)
r2_test = 1 - ss_res / ss_tot

print("\n\nTest set predictions (using training model):")
print(f"MSE: {mse_test}")
print(f"R²: {r2_test}")

# model_test = sm.OLS(y_test, X_test)
# result_test = model_test.fit()
# print(result_test.summary())
# print("\n\nMSE of testing data:")
# print(result_test.mse_resid)
# 
# print("\nP values for testing data:")
# print(result_test.pvalues)
```

```python
# Log transform features and add 1 to avoid taking log(0)
# Reuse csv instead of the standardized params
train = df_csv.drop(index=range(501, 631))
X_train_log = np.log(train.drop(columns=['Concrete compressive strength(MPa, megapascals) ']) + 1)
y_train_log = train['Concrete compressive strength(MPa, megapascals) ']

test = df_csv.iloc[501:631]
X_test_log = np.log(test.drop(columns=['Concrete compressive strength(MPa, megapascals) ']) + 1)
y_test_log = test['Concrete compressive strength(MPa, megapascals) ']

X_train_log = sm.add_constant(X_train_log)
X_test_log = sm.add_constant(X_test_log)

# Fit OLS model on log transformed training data
model_log = sm.OLS(y_train_log, X_train_log)
result_log = model_log.fit()
print(result_log.summary())
print("\n\nMSE of training data:")
print(result_log.mse_resid)

print("\nP values of training data:")
print(result_log.pvalues)

# Predict on test data using the training model
y_pred_test_log = result_log.predict(X_test_log)
mse_test_log = np.mean((y_test_log - y_pred_test_log) ** 2)
ss_res = np.sum((y_test_log - y_pred_test_log) ** 2)
ss_tot = np.sum((y_test_log - np.mean(y_test_log)) ** 2)
r2_test_log = 1 - ss_res / ss_tot

print("\n\nTest set predictions (using training model):")
print(f"MSE: {mse_test_log}")
print(f"R²: {r2_test_log}")
```
