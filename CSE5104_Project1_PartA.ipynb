```python
from ucimlrepo import fetch_ucirepo
# import json
import pandas as pd
import matplotlib.pyplot as plt 
import statsmodels.api as sm

############################################################
####################### PART A #############################
############################################################

########################################################
## Fetch dataset, Store as dataframes, Grab metadata, ##
## and Print variables/ describe original data        ##
########################################################
# concrete_compressive_strength = fetch_ucirepo(id=165) 

# X = concrete_compressive_strength.data.features 
# y = concrete_compressive_strength.data.targets 

# metaData = json.dumps(concrete_compressive_strength.metadata, indent=4)
# print(metaData) 

# print(concrete_compressive_strength.variables) 
# print(concrete_compressive_strength.data.original.describe())
  
# df = concrete_compressive_strength.data.original

#############################################################
## Displays the boxplot and summary stats for each feature ##
#############################################################
# for column in df.columns:
#     df[column].plot(kind='box')
#     plt.show()
#     print(df[column].describe())


############################################################
####################### PART C #############################
############################################################
# Let's try reading directly from the csv provided
df_csv = pd.read_csv('Concrete_Data.csv')

# Extract specified row locations for TESTING data samples and drop the DV which has a trailing space fun
test = df_csv.iloc[501:631]
X_test = test.drop(columns=['Concrete compressive strength(MPa, megapascals) '])
y_test = test['Concrete compressive strength(MPa, megapascals) ']

# Extract specified row locations for TRAINING data samples and drop the DV
train = df_csv.drop(index=range(501, 631))
X_train = train.drop(columns=['Concrete compressive strength(MPa, megapascals) '])
y_train = train['Concrete compressive strength(MPa, megapascals) ']

X_test = sm.add_constant(X_test)
X_train = sm.add_constant(X_train)

model_train = sm.OLS(y_train, X_train)
result_train = model_train.fit()
print(result_train.summary())
print("MSE of training data")
print(result_train.mse_resid)

print("P values")
print(result_train.pvalues)

model_test = sm.OLS(y_test, X_test)
result_test = model_test.fit()
print(result_test.summary())
print("MSE of testing data")
print(result_test.mse_resid)

print("P values")
print(result_test.pvalues)
```
